```python
import pandas as pd
from sqlalchemy import create_engine

# ✏️ EDIT THESE 3 LINES ONLY
CSV_PATH   = r"C:\Users\ROYAL T CONCEPT\Downloads\S&P+500+Stock+Prices+2014-2017.csv (1)\S&P 500 Stock Prices 2014-2017.csv"
MYSQL_USER = "root"
MYSQL_PASS = "farida"

# Load CSV
print("Reading CSV...")
df = pd.read_csv(CSV_PATH)

# Rename date column
df.rename(columns={"date": "trade_date"}, inplace=True)

# Fix data types
df["trade_date"] = pd.to_datetime(df["trade_date"])
df["volume"]     = pd.to_numeric(df["volume"], errors="coerce")
df["open"]       = pd.to_numeric(df["open"],   errors="coerce")
df["high"]       = pd.to_numeric(df["high"],   errors="coerce")
df["low"]        = pd.to_numeric(df["low"],    errors="coerce")
df["close"]      = pd.to_numeric(df["close"],  errors="coerce")

print(f"✅ CSV loaded: {len(df):,} rows")

# Connect to MySQL
engine = create_engine(
    f"mysql+pymysql://{MYSQL_USER}:{MYSQL_PASS}@localhost/sp500_db"
)

# Import into MySQL
print("Importing to MySQL... please wait 1-3 minutes...")
df.to_sql(
    name      = "stockanalysis_raw",
    con       = engine,
    if_exists = "replace",
    index     = False,
    chunksize = 5000
)

print("✅ DONE! All rows imported into MySQL successfully.")
```
